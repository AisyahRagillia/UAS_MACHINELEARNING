```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

import warnings

warnings.filterwarnings("ignore")
#Memanggil Library
# melakukan pemanggilan sebuah library yaitu ada pandas untuk melakukan manipulasi datanya, 
#numpy untuk melakukan perhitungannya, 
#matplotlib untuk visualisasi, 
#KMeans untuk melakukan clustering dengan metode KMeans,
#Kemudian ada library warnings yang digunakan untuk melakukan “ignore”
#ketika ada peringatan dengan kodingan yang sedang dijalankan
```

```python
data = pd.read_csv('Mall_Customers.csv')
data.head()
#memanggil data head nya
```

```python
x = data.iloc[:, [3, 4]].values
x
# codingan tersebut mengekstrak data dari kolom ke-3 dan ke-4 dari DataFrame data 
# menggunakan pandas dan menyimpannya dalam bentuk array NumPy di variabel x
```

```python
data
#menampilkan data
```

```python
from sklearn.cluster import KMeans
AB = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 44)
    kmeans.fit(x)
    AB.append(kmeans.inertia_)
    print(AB)
plt.plot(range(1, 11), AB)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('AB')
plt.show()

#untuk melakukan clustering pada data yang disimpan dalam variabel x. 
#Iterasi dilakukan untuk jumlah klaster (clusters) dari 1 hingga 10,
#dan untuk setiap iterasi, inertia (intra-cluster sum of squares)
#dari hasil clustering disimpan dalam list AB. Selanjutnya,
#hasil dari iterasi ini ditampilkan dalam bentuk grafik menggunakan matplotlib untuk membantu 
#menentukan jumlah klaster yang optimal (metode Elbow Method).
```

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 44)
y_kmeans = kmeans.fit_predict(x)
y_kmeans

#`y_kmeans` adalah variabel yang menyimpan hasil prediksi klaster (cluster labels) 
#setelah model K-Means diterapkan pada data `x`. 
#Setiap data point akan diberikan label klaster yang telah dihasilkan oleh model K-Means.
```

```python
from collections import Counter
print(Counter(y_kmeans))
#Menggunakan `Counter` dari modul `collections` 
#untuk menghitung jumlah kemunculan setiap nilai dalam array `y_kmeans`. 
#Hasilnya adalah sebuah dictionary yang menunjukkan jumlah data point
#yang terdistribusi pada setiap klaster.
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(x[y_kmeans == 3, 0], x[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(x[y_kmeans == 4, 0], x[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow',
            label = 'Centroids')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
# plt.legend()
plt.show()

#Menampilkan warna ungu dengan penghasilan rendah dengan pengeluaran tertinggi
#Menampilkan warna merah dengan penghasilan terendah dengan pengeluaran terendah
#menampilkan warna cyan dengan pengahsilan tinggi dengan pengeluaran tertinggi
#Menampilkan warna hijau dengan penghasilan tinggi dengan pengeluaran rendah
#Menampilkan warna biru dengan penghasilan sedang dengan pengeluaran sedang
```
